# Database Browser and Selector Widgets Example

This notebook demonstrates how to use the DatabaseBrowserWidget and DatabaseSelectorWidget to explore and manage database connections in JupySQL.

## Features

- **DatabaseBrowserWidget**: Browse database structure (schemas, tables, columns) in a tree view
- **DatabaseSelectorWidget**: Switch between multiple database connections easily
- **Credentials Protection**: Database passwords and sensitive information are automatically hidden

## Setup

````python
%load_ext sql
````

## Create Multiple Database Connections

Let's create a few database connections to demonstrate the widgets.

````python
# Create an in-memory SQLite database
%sql sqlite:///example1.db --alias db1
````

````python
# Create some sample data in db1
%%sql
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT NOT NULL
);

INSERT INTO users (name, email) VALUES 
    ('Alice', 'alice@example.com'),
    ('Bob', 'bob@example.com'),
    ('Charlie', 'charlie@example.com');
````

````python
# Create another database
%sql sqlite:///example2.db --alias db2
````

````python
# Create some sample data in db2
%%sql
CREATE TABLE IF NOT EXISTS products (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    price REAL NOT NULL,
    category TEXT
);

INSERT INTO products (name, price, category) VALUES 
    ('Laptop', 999.99, 'Electronics'),
    ('Mouse', 29.99, 'Electronics'),
    ('Desk', 299.99, 'Furniture');
````

## Database Browser Widget

The Database Browser Widget provides a tree view of all your database connections, schemas, tables, and columns.

**Key Features:**
- Click on connections to expand and see schemas
- Click on schemas to see tables
- Click on tables to see columns with their data types
- Credentials are automatically hidden (passwords replaced with ***)

````python
from sql.widgets import DatabaseBrowserWidget

browser = DatabaseBrowserWidget()
browser
````

## Database Selector Widget

The Database Selector Widget allows you to easily switch between different database connections.

**Key Features:**
- Dropdown showing all available connections
- Current connection is highlighted
- One-click switching between databases
- Credentials are automatically hidden

````python
from sql.widgets import DatabaseSelectorWidget

selector = DatabaseSelectorWidget()
selector
````

## Using Both Widgets Together

You can display both widgets in the same notebook for a complete database management experience.

````python
from sql.widgets import DatabaseBrowserWidget, DatabaseSelectorWidget

# Display the selector to switch connections
selector = DatabaseSelectorWidget()
display(selector)

# Display the browser to explore the database structure
browser = DatabaseBrowserWidget()
display(browser)
````

## Query After Switching

After using the selector to switch databases, you can run queries on the active connection.

````python
# This will query the currently active connection
%sql SELECT * FROM sqlite_master WHERE type='table'
````

## Advanced: Using with Real Databases

The widgets work with any database supported by JupySQL:

```python
# PostgreSQL example
%sql postgresql://user:password@localhost/mydb --alias postgres_db

# MySQL example
%sql mysql://user:password@localhost/mydb --alias mysql_db

# DuckDB example
%sql duckdb:///mydata.db --alias duckdb

# Then use the widgets
browser = DatabaseBrowserWidget()
selector = DatabaseSelectorWidget()
```

**Note:** The widgets automatically hide passwords and sensitive information, showing only `***` in place of credentials.
